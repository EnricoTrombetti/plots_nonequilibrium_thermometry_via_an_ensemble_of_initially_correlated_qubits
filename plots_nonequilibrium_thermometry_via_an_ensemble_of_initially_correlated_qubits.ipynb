```python
import matplotlib.pyplot as plt
import json
import re
import os
```

# Figure 1

```python
figure = "01"

with open("data/figure_" + figure + ".json", "r") as file:
    raw_data = file.read().strip()

#the following line replaces the scientific notation exported from Mathematica with the correct format readable by Python
fixed_data = re.sub(r'(\d)\*\^(-?\d+)', r'\1e\2', raw_data) 

data = json.loads(fixed_data)

for i in range(len(data)):
    times = []
    qfis = []
    for j in range(len(data[i])):
        time , qfi = data[i][j][0], data[i][j][1]
        times.append(time)
        qfis.append(qfi)
    plt.plot(times, qfis, marker='o')

plt.xlabel('t')
plt.ylabel('QFI')
plt.ylim(-0.1, 1.2)
plt.xlim(0,2)
plt.title('Figure ' + figure)
plt.grid(True)
plt.show()
```

```python
figure = "01_inset"

with open("data/figure_" + figure + ".json", "r") as file:
    raw_data = file.read().strip()

#the following line replaces the scientific notation exported from Mathematica with the correct format readable by Python
fixed_data = re.sub(r'(\d)\*\^(-?\d+)', r'\1e\2', raw_data) 

data = json.loads(fixed_data)

for i in range(len(data)):
    times , negs = data[i][0], data[i][1]
    plt.plot(times, negs, marker='o')

plt.xlabel('t')
plt.ylabel('Negativity')
plt.ylim(0, 0.6)
plt.xlim(0,0.3)
plt.title('Figure ' + figure)
plt.grid(True)
plt.show()
```
