```python
import matplotlib.pyplot as plt
import json
```

# Figure 1

```python
figure = "01"

with open("data/figure_" + figure + ".json", "r") as file:
    raw_data = file.read().strip()

data = json.loads(raw_data)

for i in range(len(data)):
    times , qfis = data[i][0], data[i][1]
    plt.plot(times, qfis, marker='o')
    

plt.xlabel('t')
plt.ylabel('QFI')
plt.ylim(-0.1, 1.3)
plt.xlim(0,2)
plt.title('Figure ' + figure)
plt.grid(True)
plt.show()
```

```python
figure = "01_inset"

with open("data/figure_" + figure + ".json", "r") as file:
    raw_data = file.read().strip()

data = json.loads(raw_data)

for i in range(len(data)):
    times , negs = data[i][0], data[i][1]
    plt.plot(times, negs, marker='o')

plt.xlabel('t')
plt.ylabel('Negativity')
plt.ylim(0, 0.6)
plt.xlim(0,0.3)
plt.title('Figure ' + figure)
plt.grid(True)
plt.show()
```
