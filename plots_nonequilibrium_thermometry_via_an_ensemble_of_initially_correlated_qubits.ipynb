```python
import matplotlib.pyplot as plt
import json
import re
```

# Figure 2

```python
figure = "02a"

with open("data/figure_" + figure + ".dat", "r") as file:
    raw_data = file.read().strip()

#the following line replaces the scientific notation exported from Mathematica with the correct format readable by Python
fixed_data = re.sub(r'(\d)\*\^(-?\d+)', r'\1e\2', raw_data) 

data = json.loads(fixed_data)

for i in range(len(data)):
    x , y = data[i][0], data[i][1]
    plt.plot(x, y, marker='o')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Figure ' + figure)
plt.grid(True)
plt.show()
```

# Figure 4

```python
figure = "04"

with open("data/figure_" + figure + ".dat", "r") as file:
    raw_data = file.read().strip()

#the following line replaces the scientific notation exported from Mathematica with the correct format readable by Python
fixed_data = re.sub(r'(\d)\*\^(-?\d+)', r'\1e\2', raw_data) 
fixed_data = re.sub(r'(?<=\d)\.(?!\d)', '.0', fixed_data)

data = json.loads(fixed_data)

for i in range(len(data)):
    x_vals=[]
    y_vals=[]
    for j in range(len(data[i])):
        x_vals.append(data[i][j][0])
        y_vals.append(data[i][j][1])
    plt.plot(x_vals, y_vals, marker='o')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Figure ' + figure)
plt.grid(True)
plt.show()
```
